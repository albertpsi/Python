# Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series
## A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd
# Create a Series from a list
data = [1,2,3,4,5]
series = pd.Series(data) # index will be automatically created as 0,1,2,3,4
print(f"Series from list:\n{series}\n")
type(series)
```

```python
# Create a series from a dictionary
data_dict = {'a': 1, 'b': 2, 'c': 3}
series_dict = pd.Series(data_dict) # keys becomes index and values becomes data
print(f"Series from dictionary:\n{series_dict}\n")
type(series_dict)
```

```python
data = [10,20,30,40,50]
index = ['a','b','c','d','e']
series_index = pd.Series(data, index=index) # index will be created as a,b,c,d,e
print(f"Series with custom index:\n{series_index}\n")
```

```python
# DataFrame
# A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). 
# It is similar to a spreadsheet or SQL table.

# Create a DataFrame from a dictionary of lists
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Tokyo']
}
df = pd.DataFrame(data_dict)
print(f"DataFrame from dictionary:\n{df}\n")
type(df)


import numpy as np
np.array(df)
```

```python
# create a DataFrame from a list of dictionaries
data_list = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'London'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Tokyo'}
]
df = pd.DataFrame(data_list)
print(f"DataFrame from list of dictionaries:\n{df}\n")
```

```python
# Accessing data from DataFrame
data_list = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'London'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Tokyo'}
]
df = pd.DataFrame(data_list)
df
```

```python
df['Name'] # Accessing a column
```

```python
df[['Name', 'City']] # Accessing multiple columns
```

```python
df.loc[0, 'Name'] # Accessing a specific cell by row and column labels
```

```python
df.iloc[0] # Accessing a row by position
```

```python
# Difference between loc and iloc
# loc is label-based, which means that you have to specify the name of the row or column that you want to access. 
# iloc is integer position-based, which means that you have to specify the index of the row or column that you want to access.
```

```python
# Accessing a specified element
print(f"Age at row 1: {df.at[1,'Age']}") # Accessing a specific cell by row and column labels
```

```python
# Accessing a specified element using iat
print(f"Age at row 1: {df.iat[1,1]}") # Accessing a specific cell by row and column position
```

```python
# Data Manupulation with DataFrame

df
```

```python
# Adding a new column to the DataFrame
df['Salary'] = [50000, 60000, 70000] # Adding a new column
df
```

```python
# remove a column from the DataFrame
df.drop('Salary', axis=1, inplace=True) # Removing a column for column we have to specify axis=1, inplace=True will make the change in the original DataFrame
# to remove a row we have to specify axis=0
# inplace=True will make the change in the original DataFrame, 
# if inplace=False then it will return a new DataFrame with the column removed
df
```

```python
df['Age'] = df['Age'] + 1 # Accessing a column
df
```

```python
# drop based on row index
df.drop(0, axis=0, inplace=True) # Removing a row for row we have to specify axis=0, 
# inplace=True will make the change in the original DataFrame
df
```
