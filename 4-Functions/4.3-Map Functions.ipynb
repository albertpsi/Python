# The map() Function in Python
The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

## Map Function Syntax
map(function, iterable, ...)


```python
def square(num):
    return num*num

square(2)
```

```python
# abobe func will fail if we pass list or set etc,
square([1,2,3,4]) # Type Error
```

```python
# so we will use map function

map(square,[1,2,3,4])  # creates a map object at memory loc
# convert it into a list

list(map(square,[1,2,3,4]))
```

```python
## Lambda function with map
numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
```

```python
### MAp multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
```

```python
# map to convert list of strings to integers

str_num =['1','2','3','4','5']

list(map(lambda x:int(x),str_num))
```
