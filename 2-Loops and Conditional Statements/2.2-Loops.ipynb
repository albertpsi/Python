# Loops

```python
# for loop - Traversing a Collection

for i in range(5):
    print("Iteration:", i)
```

```python
# Range Function
# The range() function generates a sequence of numbers, which is commonly used in for loops.
# It can take one, two, or three arguments: start, stop, and step.
# - If one argument is provided, it is treated as the stop value, and the sequence starts from 0.
# - If two arguments are provided, the first is the start value, and the second is the stop value.
# - If three arguments are provided, the third is the step value, which determines the increment between each number in the sequence.
# Last Number in stop value is Excluded

# Example with start, stop, and step
for i in range(1, 10, 2):
    print("Odd Number:", i) #Prints odd numbers from 1 to 9, excluding 10

# Example with negative step
for i in range(10, 0, -2):
    print("Countdown Even Number:", i) # excludes 0

# Example without step (default is 1)
for i in range(3, 8):
    print("Number:", i) #Prints numbers from 3 to 7, excluding 8
```

```python
# string iteration
str = 'Sai Sampath'

for i in str:
    print(i)
```

```python
# While Loop
# A while loop repeatedly executes a block of code as long as a specified condition is True.
count = 0
while count < 5:
    print("Count is:", count)
    count += 1  # Increment the counter to avoid infinite loop
```

```python
# Loop Control Statements
# Loop control statements are used to modify the behavior of loops. The main loop control statements in Python are:
# 1. break: Terminates the loop entirel, exits loop prematurely. 
# 2. continue: Skips the current iteration and moves to the next iteration of the loop.
# 3. pass: A placeholder that does nothing; it is used when a statement is syntactically required but no action is needed.

for i in range(5):
    if i == 3:
        break  # Exit the loop when i is 3
    print("Break Example - i:", i)
```

```python
# Continue Example

for i in range(5):
    if i == 2:
        continue  # Skip the iteration when i is 2
    print("Continue Example - i:", i)
# Pass Example
for i in range(5):
    if i == 4:
        print(f"Pass Example - i is {i}, doing nothing")
        pass  # Do nothing when i is 4
    print("Pass Example - i:", i)
```

```python
# Nested Loops
# A nested loop is a loop inside another loop. The inner loop runs completely for each iteration of the outer loop.
for i in range(3):  # Outer loop
    for j in range(2):  # Inner loop
        print(f"Nested Loop - i: {i}, j: {j}")
```

```python
# Sum of Natural Numbers using while loop and  for loop

count = 0
sum = 0
num = int(input("Enter a positive integer: "))

while count <= num:
    sum += count
    count += 1
print(f"The sum of natural numbers up to {num} is: {sum}")

# Using for loop
sum = 0
for i in range(0,num + 1):
    sum += i
print(f"The sum of natural numbers up to {num} using for loop is: {sum}")
```

```python
# Prime Number Checker
num = int(input("Enter a positive integer: "))
is_prime = True
if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
 # Prime Numbers between a Range
lower = int(input("Enter the lower bound: "))
upper = int(input("Enter the upper bound: "))
print(f"Prime numbers between {lower} and {upper} are:")
for num in range(lower, upper + 1):
    is_prime = True
    if num <= 1:
        is_prime = False
    else:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
    if is_prime:
        print(num) 
```
