# Functions
A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
## syntax
def function_name(parameters):
    """Docstring"""  # comments on what func does - optional 
    # Function body
    return expression # optional some func can be without return
```

```python
# return true if the num is even

def isEven(num):
    """This function finds even or odd"""
    if(num %2 ==0):
        return True
    else:
        return False

isEven(24)  # calling the func 
```

```python
# A func to add 2 numbers

def add(a,b):
    return a+b

result = add(2,4)
print(result)
```

```python
# Default Parameters

def greet(name = "Guest"):
    print(f"Hello {name} Welcome to the Paradise!")

greet("Sai")
greet() # will use default parameter
```

```python
# Variable  Length Arguments
# Positional and Keywords Arguments
def variable_args(*args, **kwargs):
    print("Positional arguments (*args):", args) # * - is for positional args
    print("Keyword arguments (**kwargs):", kwargs) # ** - double * is for keywords args, in keyword args all the args will be in key-value paired
variable_args(1, 2, 3, 4, "Sai", name="Alice", age=30) # this will work fine
variable_args(1, 2, 3, 4, "Sai", name="Alice", age=30, "Sampath") # this will give an error because the positional args should be before the keyword args
```

```python
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result, len(args)  # func can return multiple values, here we are returning the product and the count of numbers

print(multiply(2, 3, 4)) # Output: (24, 3)
```

```python
# Temperature conversion func

def convertTemperature(temp,unit):
    """This function converts tempeprature from Celsius to Fahrenheit and vice versa"""
    if unit == "C":
        return (temp - 32) * 5.0/9.0, "F"
    elif unit == "F":
        return (temp * 9.0/5.0) + 32, "C"
    else:
        return "Invalid unit! Please use 'C' for Celsius or 'F' for Fahrenheit."
    
print(convertTemperature(100, "C")) # Output: (37.77777777777778, 'F')
print(convertTemperature(212, "F")) # Output: (100.0, 'C')
print(convertTemperature(50, "X")) # Output: Invalid unit! Please use 'C' for Celsius or 'F' for Fahrenheit.
```

```python
# Password Strength Checker
def checkPasswordStrength(password):
    """This function checks the strength of a password based on its length and character variety."""
    if len(password) < 6:
        return "Weak"
    elif len(password) < 12:
        return "Moderate"
    else:
        return "Strong"
```

```python
# calculate the total cost of items in shopping cart
def calculateTotalCost(**kwargs):
    """This function calculates the total cost of items in a shopping cart."""
    total_cost = 0
    for item, price in kwargs.items():
        if not isinstance(price, (int, float)):
            print(f"Warning: The price of {item} is not a valid number. Skipping this item.")
            continue
        total_cost += price
    return total_cost

print(calculateTotalCost(apple=1.5, bread=2.0, milk=3.0, iphone = 999.99)) # Output: 1006.5
```

```python
# check if a string is a palindrome
def isPalindrome(s):
    """This function checks if a string is a palindrome."""
    cleaned_string = ''.join(s.split()).lower()  # Remove spaces and convert to lowercase
    return cleaned_string == cleaned_string[::-1]  # Check if the string is equal to its reverse

print(isPalindrome("A man a plan a canal Panama")) # Output: True
print(isPalindrome("Hello")) # Output: False
```

```python
#A Function To Read A File and count the frequency of each word
def count_word_frequency(file_path):
    word_count={}
    with open(file_path,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip('.,!?;:"\'')
                word_count[word]=word_count.get(word,0)+1
    
    return word_count

filepath='sample.txt'
word_frequency=count_word_frequency(filepath)
print(word_frequency)
```
