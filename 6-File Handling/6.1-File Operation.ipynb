# File Operation- Read And Write Files
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
# read a whole file
with open('example.txt', 'r') as file: # open the file in read mode and assign it to the variable 'file'
    content = file.read() # read the entire contents of the file and store it in the variable 'content'
    print(content)
```

```python
with open('example.txt', 'r') as file: # open the file in read mode and assign it to the variable 'file'
    for line in file: # iterate over each line in the file
        print(line.strip()) # print the line after stripping leading and trailing whitespace characters, removes new line characters as well
```

```python
with open('example.txt', 'w') as file: # open the file in write mode and assign it to the variable 'file'
    file.write("This is a new line of text.\n") # write a new line of text to the file, including a newline character at the end
    file.write("This is another line of text.\n") # write another line of text to the file, including a newline character at the end

    # w - write mode - overwite the file contents
    # a - append mode - add new content to the end of the file without overwriting existing content
with open('example.txt', 'a') as file: # open the file in append mode and assign it to the variable 'file'
    file.write("This line will be appended to the file.\n") # write a new line of text to the end of the file, including a newline character at the end    
    
```

```python
# writing a list of lines to a file
lines = ["First line of text.\n", "Second line of text.\n", "Third line of text.\n"] # create a list of lines to be written to the file, each line ends with a newline character
with open('example.txt', 'a') as file: # open the file in append mode
    file.writelines(lines) # write all lines in the list to the file
```

```python
# Binary file handling - Bin File

with open('example.bin', 'wb') as file: # open a binary file in write bites mode
    file.write(b'\x00\x01\x02\x03\x04') # write binary data to the file - the b prefix indicates that the data is in bytes format
```

```python
with open('example.bin', 'rb') as file: # open the file in read bites mode and assign it to the variable 'file'
    for line in file: # iterate over each line in the file
        print(line.strip()) # print the line after stripping leading and trailing whitespace characters
```

```python
# Read content from a source text file and write into a destination text file
with open('example.txt','r') as sourceFile:
    content = sourceFile.read() # read the entire contents of the source file and store it in the variable 'content'
    with open('destination.txt', 'w') as destFile: # open the destination file in write mode
        destFile.write(content) # write the content read from the source file into the destination file - if file does not exist, it will be created; if it already exists, its contents will be overwritten

with open('destination.txt', 'r') as file: # open the destination file in read mode
    print(file.read()) # read and print the contents of the destination file to verify that the content was copied correctly
```

```python
# writing and then reading a file using with statement
with open('example.txt', 'w+') as file: # open the file in write+ mode (allows both writing and reading)
    file.write("adding this text using w+ mode i.e write and read mode. write will overwrite existing content\n") # write a line of text to the file
    # After writing, the file pointer is at the end of the file, so we need to move it back to the beginning before reading
    file.seek(0) # move the file pointer to the beginning of the file
    
    content = file.read() # read the contents of the file and store it in the variable 'content'
    print(content) # print the content to verify that it was written and read correctly
```
