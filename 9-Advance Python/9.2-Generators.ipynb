# Generators
Generators are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
# Generators

# function fo calculate the square of a number
def square(n):
    for i in range(n):
        yield i ** 2 # yield is used to return a generator object instead of a list. 
        #It allows us to iterate over the values one at a time, which can be more memory efficient for large datasets.
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
square = square(3)
print(next(square)) #0
print(next(square)) #1
print(next(square)) #4
```

```python
iterator = iter(square(3))
print(next(iterator)) #0
print(next(iterator)) #1
print(next(iterator)) #4
```

## Generators and Large Files
Generators are used to read large files line by line without storing the entire file in memory.

```python
def read_large_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            yield line.rstrip("\n")

for line in read_large_file("large_file.txt"):
    print(line)
```
