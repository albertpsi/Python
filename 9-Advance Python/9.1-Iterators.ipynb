# Iterators
Iterators are advanced Python concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.


````python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
````

````python
iterator = iter(my_list)
print(type(iterator))
iterator #<list_iterator object at 0x7f8c8c8c8c8c>, it is an object of type list_iterator
# It uses lazy evaluation, it does not compute the next value until it is requested
print(next(iterator)) #1
print(next(iterator)) #2
````

````python
my_set = {1,2,3,4,5,6}
iterator_set = iter(my_set)
print(type(iterator_set))
print(next(iterator_set)) #1
````

````python
my_tuple=(1,2,3,4,5,6)
iterator_tuple = iter(my_tuple)
print(type(iterator_tuple))
print(next(iterator_tuple)) #1
````

````python
print(next(iterator_tuple)) #2
print(next(iterator_tuple)) #3
print(next(iterator_tuple)) #4
print(next(iterator_tuple)) #5
print(next(iterator_tuple)) #6
print(next(iterator_tuple)) #StopIteration, it raises an error because there are no more elements to iterate over
````

````python
my_dict = {'a':1, 'b':2, 'c':3}
iterator_dict = iter(my_dict)
print(type(iterator_dict))
print(next(iterator_dict)) #a, it iterates over the keys of the dictionary, not the values
print(next(iterator_dict)) #b
````

## Using Iterators When Collection Size Is Unknown

That is exactly what iterators are for.

You do not need to know the size up front. You pull values one by one until data ends (or until you decide to stop).

```python
it = iter(data_source)

while True:
    try:
        item = next(it)   # get next item
        process(item)
        if should_stop(item):   # optional custom stop condition
            break
    except StopIteration:
        break
```

In practice, `for` is cleaner and handles `StopIteration` for you:

```python
for item in data_source:
    process(item)
```

For very large or infinite streams, limit safely with `itertools.islice`:

```python
from itertools import islice

for item in islice(data_source, 100):   # only first 100 items
    process(item)
```

So iterators let you work with unknown-size (even unbounded) collections efficiently, one element at a time.
