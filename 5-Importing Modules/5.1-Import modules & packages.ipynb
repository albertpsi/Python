# Importing Modules in Python: Modules and Packages
In Python, modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math # import math package
math.sqrt(16)
```

```python
from math import sqrt, pi # imports sqrt and pi functions from math package
print(f"Square roots of 16 and 25 are: {sqrt(16)}, {sqrt(25)}")
print(pi)
```

```python
import numpy as np # import numpy package and give it an alias np
print(np.array([1, 2, 3]))
print(type(np.array([1, 2, 3])))
```

```python
from math import * # imports all functions from math package
print(sqrt(16))
```

```python
from package import maths # imports the maths module from package
print(maths.addition(5, 3))
```

```python
from package.maths import addition # imports the maths module from package
print(addition(5, 3))
```
