# Classes and Objects

Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects. This lesson covers the basics of creating classes and objects, including instance variables and methods.

```python
### A class is a blue print for creating objects. Attributes,methods
class Car:
    pass

audi=Car()
bmw=Car()

print(type(audi))
print(type(bmw))

print(audi)
print(bmw)

print(audi==bmw)
```

```python
# Instance variable and methods
class Dog:
    # Constructor
    def __init__(self,name,age):
        self.name=name # self.name is instance variable similar to this.name in C#
        self.age=age
    def bark(self):
        print("Woof! Woof!")

dog1=Dog("Buddy",3)
dog2=Dog("Max",5)

print(dog1) # prints the memory address of the object
print(dog1.name) # prints the name of the dog
print(dog1.age) 
dog2.bark() # prints the bark sound


print(dog2.bark()) # prints the bark sound and then prints None
'''dog2.bark() already does a print("Woof! Woof!"), but the method does not return anything.
In Python, a function/method with no return implicitly returns None.

So execution is:

dog2.bark() prints Woof! Woof!
dog2.bark() returns None
outer print(...) prints that None'''
```

```python
# Modeling a Bank Account
# Define a class BankAccount with attributes account_number, balance and methods deposit, withdraw and display_balance
```

```python
class BankAccount:
    def __init__(self,owner,account_number,balance=0):
        self.owner=owner
        self.account_number=account_number
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposited {amount}. New balance: {self.balance}")
    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficient funds")
        else:
            self.balance-=amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
    def display_balance(self):
        print(f"Account Number: {self.account_number}, Balance: {self.balance}")
```

```python
deposit_amount=int(input("Enter amount to deposit: "))
withdraw_amount=int(input("Enter amount to withdraw: "))

account=BankAccount("Alice","123456789",1000)
account.display_balance()
account.deposit(deposit_amount)
account.withdraw(withdraw_amount)
account.display_balance()
```
