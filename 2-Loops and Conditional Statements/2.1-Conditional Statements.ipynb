```python
## if statement
# if statement is used to execute a block of code when a certain condition is True
age=20

if age>=18:
    print("You are allowed to vote in the elections")
```

```python
## else statement
# else statement is used to execute a block of code when the if condition is False
age=16
if age>=18:
    print("You are allowed to vote in the elections")   
else:
    print("You are not allowed to vote in the elections as you are minor")
```

```python
## elif statement
# elif statement is used to check multiple conditions. It stands for "else if"
age = int(input("Enter your age: "))
if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
## Nested Conditional Statements
# Nested conditional statements are if-elif-else statements inside another if-elif-else statement.

num = int(input("Enter a number:"))

if num > 0:
    print("The number is positive.")
    if num % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is zero or negative.")
```

```python
#Leap Year Checker
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
# Common Mistakes to Avoid
# 1. Indentation Errors: Ensure proper indentation for blocks of code and use : after conditions.
# 2. Using Assignment Operator Instead of Comparison Operator: Use '==' for comparison, not '='.
# 3. Forgetting to Cover All Conditions: Use else to cover all possible cases.
# 4. Overcomplicating Conditions: Keep conditions simple and readable.
# 5. Not Testing Edge Cases: Test your conditions with edge cases to ensure correctness.
```
