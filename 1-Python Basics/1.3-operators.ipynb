# Deep Dive into Operators

### 1 .Arithmetic Operators
    Addition
    Subtraction
    Multiplication
    Division
    Floor Division
    Modulus
    Exponentiation

```python
a = 10
b = 20
sum_result = a + b
print(f"Sum of {a} and {b} is {sum_result}")  # Output: Sum of 10 and 20 is 30
```

```python
a = 10
b = 20
diff_result = b - a
print(f"Difference between {b} and {a} is {diff_result}")  # Output: Difference between 20 and 10 is 10
prod_result = a * b
print(f"Product of {a} and {b} is {prod_result}")  # Output: Product of 10 and 20 is 200
div_result = b / a  
print(f"Division of {b} by {a} is {div_result}")  # Output: Division of 20 by 10 is 2.0
floor_div_result = b // a
print(f"Floor Division of {b} by {a} is {floor_div_result}")  # Output: Floor Division of 20 by 10 is 2
mod_result = 20 % 3
print(f"Modulus / Reminder of {20} by {3} is {mod_result}")  # Output: Modulus of 20 by 10 is 0
exp_result = a ** 2
print(f"{a} raised to the power 2 is {exp_result}")  # Output: 10 raised to the power 2 is 100
```

```python
# Diff B/w Division and Floor Division
a = 7
b = 3
div_result = a / b
floor_div_result = a // b
print(f"Division of {a} by {b} is {div_result}")  # Output: Division of 7 by 3 is 2.3333333333333335
print(f"Floor Division of {a} by {b} is {floor_div_result}")  # Output: Floor Division of 7 by 3 is 2  
# Floor division returns the largest integer less than or equal to the division result, 
# i.e it truncates the decimal part and rounds down to the nearest whole number. 
```

# Comparison Operators
    Equal to
    Not equal to
    Greater than
    Less than
    Greater than or equal to
    Less than or equal to

```python
# Comparison Operators
a = 10
b = 20
print(f"{a} == {b}: {a == b}")  # Output: 10 == 20: False
print(f"{a} != {b}: {a != b}")  # Output: 10 != 20: True
print(f"{a} > {b}: {a > b}")    # Output: 10 > 20: False
print(f"{a} < {b}: {a < b}")    # Output: 10 < 20: True
print(f"{a} >= {b}: {a >= b}")  # Output: 10 >= 20: False
print(f"{a} <= {b}: {a <= b}")  # Output: 10 <= 20: True
```

```python
# Comparison Operators with Strings
name = "Sai"
name_to_check = "sai"
print(f'"{name}" == "{name_to_check}": {name == name_to_check}')  # Output: "Sai" == "sai": False # Because of case sensitivity
print(f'"{name}" != "{name_to_check}": {name != name_to_check}')  # Output: "Sai" != "sai": True
print(f'"{name}" > "{name_to_check}": {name > name_to_check}')    # Output: "Sai" > "sai": False # Because uppercase 'S' has lower ASCII value than lowercase 's'
print(f'"{name}" < "{name_to_check}": {name < name_to_check}')    # Output: "Sai" < "sai": True # Because uppercase 'S' has lower ASCII value than lowercase 's'
print(f'"{name}" >= "{name_to_check}": {name >= name_to_check}')  # Output: "Sai" >= "sai": False # Because uppercase 'S' has lower ASCII value than lowercase 's'
print(f'"{name}" <= "{name_to_check}": {name <= name_to_check}')  # Output: "Sai" <= "sai": True # Because uppercase 'S' has lower ASCII value than lowercase 's'
```

# Logical Operators
    AND
    OR
    NOT

```python
# And , Or , Not Operators
a = True
b = False   
c=True
print(f"{a} and {b}: {a and b}")  # Output: True and False: False
print(f"{a} and {c}: {a and c}")  # Output: True and True: True
print(f"{a} or {b}: {a or b}")    # Output: Or Operator True or False: True
print(f"not {a}: {not a}")        # Output: Not Operator - not True: False # inverses the value
print(f"not {b}: {not b}")        # Output: Not Operator - not False: True # inverses the value
```
