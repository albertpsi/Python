```python
import logging
# create a logger for module one

logger_one = logging.getLogger('module_one')
logger_one.setLevel(logging.DEBUG)

# create a logger for module two
logger_two = logging.getLogger('module_two')
logger_two.setLevel(logging.WARNING)

# configure logger
logging.basicConfig(
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', 
    level=logging.DEBUG,
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
# log messages from module one
logger_one.debug('This is a debug message from module one')

# log messages from module two
logger_two.warning('This is a warning message from module two')    
```
