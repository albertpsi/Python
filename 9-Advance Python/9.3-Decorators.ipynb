# Decorators
Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

```python
# Function copy
def welcome():
    return "Welcome, Sai!"

welcome()  # Copying the function reference
```

```python
wel = welcome  # Copying the function reference
wel()  # Calling the copied function reference
del welcome   # Deleting the original function reference
```

```python
wel() # Calling the copied function reference after deleting the original
```

```python
welcome()  # This will raise an error since 'welcome' has been deleted
```

```python
# Closures - function inside a function
def outer(msg):
    message = msg  # This variable is captured by the inner function (closure)
    
    def inner():
        print(message)  # Accessing the outer variable, parameters, and functions
    
    return inner()  # Returning the inner function

outer("Hello, Sai!") # Calling the outer function which in turn calls the inner function
```

```python
# Decorators - A decorator is a function that takes another function as an argument, 
# adds some functionality to it, and returns a new function with the added functionality.

```

```python
def main_welcome(func): # func is a parameter that will accept a function reference
   
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method ()
```

```python
def course_intro():
    print("This is an introduction to the course.")
```

```python
main_welcome(course_intro)
```

```python
@main_welcome # This is a decorator syntax, it is equivalent to main_welcome(course_intro)
def course_intro():
    print("This is an introduction to the course.")

```

```python
course_intro() # This will give an error because in main_welcome 
# we are not returning the sub_welcome_method, we are calling it. 
# So when we call course_intro() it will try to call the sub_welcome_method which is not returned from main_welcome, 
# hence it will give an error.
```

```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
# Decorators with arguments

# A decorator that takes an argument, in this case, 
# the number of times to repeat the function call.

def repeat(n): # this is the actual decorator function that takes the number of times to repeat as an argument
    def decorator(func): #this is the closure function that will be returned by the repeat function, it takes the function to be decorated as an argument
        def wrapper(*args, **kwargs): # this is a closuer function that will be returned by the decorator, it takes any number of arguments and keyword arguments
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello!")

say_hello()
```
