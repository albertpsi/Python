# Reading Data From Different Sources


```python
import pandas as pd
from io import StringIO # Importing StringIO to read JSON data from a string
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df=pd.read_json(StringIO(Data))
df
```

```python
df.to_json(orient='records') # Converting the DataFrame to JSON format with 'records' orientation 
# What is records orientation in pandas to_json method?
# The 'records' orientation in the pandas to_json method formats the JSON output as a list of dictionaries, 
# where each dictionary represents a row in the DataFrame. 
# Each key in the dictionary corresponds to a column name, and the value is the data for that column in that row. 
# This format is particularly useful for representing tabular data in JSON format, making it easy to read and work with in various applications.
```

```python
url="https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"

df = pd.read_html(url)
```

```python
df # why is it printed as a list?
# The output is printed as a list because the pd.read_html() function returns a list of DataFrames. 
# Each DataFrame in the list corresponds to a table found in the HTML content of the specified URL.
# In this case, the URL contains multiple tables, and pd.read_html() has extracted all of them, resulting in a list of DataFrames. 
# If you want to access a specific table, you can index the list, for example, df[0] to access the first table.
```

```python
url="https://en.wikipedia.org/wiki/Mobile_country_code"
pd.read_html(url,match="Country",header=0)[0]
# Reading tables from a Wikipedia page, matching the table that contains the header "Country" and using the first row as the header for the DataFrame.
```

pd.read_html(url) is hitting Wikipedia directly with pandas’ default request, and Wikipedia is returning 403 Forbidden because that request looks too “bot-like”.

```python
pd.read_html(
    url,
    match="Country",
    header=0,
    storage_options={"User-Agent": "Mozilla/5.0"},
)[0]
# Reading tables from a Wikipedia page, matching the table that contains the header "Country" and using the first row as the header for the DataFrame, 
# while also specifying a custom User-Agent in the storage options to avoid potential issues with web scraping.
```
