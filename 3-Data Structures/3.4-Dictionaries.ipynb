# Dictionaries
Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
# Creating a Dictionary

empty_dict = {} # List - [], tuple - (), Set - {}
print(type(empty_dict))
print(type(dict()))

# Creating a Dictionary with values

dict = {"name": "John", "age": 30, "city": "New York"}
print(type(dict))
```

```python
student = {"name":"Sai","age":21,"gender":"Male"}
print(student)
print(type(student))

# Accessing Dictionary Values

print(student["name"]) # can access using key - dict[key]
print(student.get("age")) # can access using key - dict.get(key)
print(student.get("gender"))

# Keys are unique

print(student.get("last_name","Not Found")) # when key is not found it returns default value
```

```python
# Modifying a Dictionary
## Dictionary is mutable, we can add, remove, update elements but key needs to be unique

print(student)
# Get Keys
print(student.keys()) # Returns all keys
# Get Values
print(student.values()) # Returns all values
# Get Items
print(student.items()) # Returns all items


# Adding new key-value pair
student["course"] = "Computer Science" # If key is not present, it will add new key-value pair, if present it will update the value
student["address"] = "123 Main St"
student["phone"] = "123-456-7890"
student["grade"] = "A"
print(student)
```

```python
# Delete key
del student["grade"] # delete the key and value pair
print(student)
```

```python
student={'name': 'Sai', 'age': 21, 'gender': 'Male', 'course': 'Computer Science', 'address': '123 Main St', 'phone': '123-456-7890'}
keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()  # returns list of tuples
print(items)
```

```python
# Shallow Copy
student={'name': 'Sai', 'age': 21, 'gender': 'Male', 'course': 'Computer Science', 'address': '123 Main St', 'phone': '123-456-7890'}
student_copy = student
print(student)
print(student_copy)

student["name"]="sampath"
print("After update")
print(student)
print(student_copy)

print("Both Student and Student_copy's name key will be changed to sampath, as both are pointing to the same memory location")

```

```python
# For Shallow Copy We will use copy() function

student={'name': 'Sai', 'age': 21, 'gender': 'Male', 'course': 'Computer Science', 'address': '123 Main St', 'phone': '123-456-7890'}
student_copy = student.copy()
print(student)
print(student_copy)

student["name"]="sampath"
print("After update")
print(student)
print(student_copy)

# Value of name key in Student_copy wont change to Sampath as it is a shallow copy, both are pointing to different memory location
```

```python
# Iterating over Dictionary
# iterate over keys

student={'name': 'Sai', 'age': 21, 'gender': 'Male', 'course': 'Computer Science', 'address': '123 Main St', 'phone': '123-456-7890'}

for keys in student.keys():
    print(keys)

# iterate over values
print("---------------values----------------------")
for values in student.values():
    print(values)

# iterate over items
print("---------------items----------------------")
for items in student.items():
    print(items)

# Iterate over key value pair
print("---------------using dict-------------------")
for key, value in student.items():
    print(f"{key}: {value}")

```

```python
# Nested Dictionaries

## Nested Disctionaries
students={
    "student1":{"name":"Peter","age":32},
    "student2":{"name":"Parker","age":35}
}
print(students)

## Access nested dictionaries elementss
print(students["student2"]["name"])
print(students["student2"]["age"])

students.items()

## Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## USe a dictionary to count he frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
