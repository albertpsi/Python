# Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns # seaborn is built on top of matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics.
```

```python
# basic plotting with seaborn
sns.load_dataset('tips') # load the 'tips' dataset tips is a built-in dataset in seaborn that contains information about restaurant tips, including the total
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=sns.load_dataset('tips')) # create a scatter plot of total bill vs tip
plt.title("Scatter Plot of Total Bill vs Tip") # add a title to the plot
plt.xlabel("Total Bill") # label the x-axis
plt.ylabel("Tip") # label the y-axis
plt.show() # display the plot
```

```python
# Line plot
sns.lineplot(x='size', y='total_bill', data=sns.load_dataset('tips')) 
# create a line plot of size vs total bill
plt.title("Line Plot of Size vs Total Bill") # add a title to the plot
plt.xlabel("Size") # label the x-axis
plt.ylabel("Total Bill") # label the y-axis
plt.show() # display the plot
```

```python
# categorical plot
sns.barplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
```

```python
# Box plot - a box plot is a standardized way of displaying the distribution of data based on a 
# five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 
# It can also show outliers in the data.
sns.boxplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
```

```python
# violin plot - a violin plot is a method of plotting numeric data and can be understood as a combination of a box plot and a kernel density plot.
sns.violinplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
```

```python
# Histogram
sns.histplot(sns.load_dataset('tips')['total_bill'], bins=20, kde=True) 
# create a histogram of total bill with 20 bins and a kernel density estimate
# kde stands for kernel density estimate, which is a way to estimate the probability density function of a continuous random variable.
plt.title("Histogram of Total Bill") # add a title to the plot
plt.xlabel("Total Bill") # label the x-axis
plt.ylabel("Frequency") # label the y-axis
plt.show() # display the plot
```

```python
# kde plot - a kernel density estimate (KDE) plot is a way to visualize the distribution of a dataset. It is a smoothed version of a histogram and can be used to estimate the probability density function of a continuous random variable.
sns.kdeplot(sns.load_dataset('tips')['total_bill'], shade=True)
```

```python
# pair plot - a pair plot is a grid of scatter plots that shows the relationships between multiple variables in a dataset. It is useful for visualizing the pairwise relationships between variables and can help identify patterns and correlations in the data.
sns.pairplot(sns.load_dataset('tips'))
```

```python
# corr - a correlation matrix is a table showing the correlation coefficients between many variables. 
# Each cell in the table shows the correlation between two variables. 
# The value is in the range of -1 to 1. 
# If the value is close to 1, it means there is a strong positive correlation between the two variables. 
# If the value is close to -1, it means there is a strong negative correlation between the two variables. 
# If the value is close to 0, it means there is no correlation between the two variables.
sns.load_dataset('tips')[['total_bill', 'tip', 'size']].corr()
```

```python
# heatmap - a heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. It is often used to visualize the correlation between variables in a dataset.
sns.heatmap(sns.load_dataset('tips')[['total_bill', 'tip', 'size']].corr(), annot=True, cmap='coolwarm')
# annot=True adds the correlation coefficients to the heatmap, and cmap='coolwarm' sets the color scheme of the heatmap.
```
