# Lambda Functions in Python
Lambda functions are small anonymous functions defined using the lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.

```python
#Syntax
lambda arguments: expression
```

## C# vs Python Lambda
In C#, lambda expressions are written as `(...) => ...`. In Python, lambda syntax is `lambda arguments: expression`.

Python `lambda` is closest to a C# anonymous function assigned to a delegate (`Func<>`/`Action<>`).

Example:
`Func<int, int> f = x => x + 1;` (C#)
`f = lambda x: x + 1` (Python)

Key difference: Python `lambda` supports only a single expression, while C# lambdas can also use statement blocks (`=> { ... }`).

```python
# lambda function that adds 2 arguments
add = lambda x,y : x + y
print(type(add)) # Output: <class 'function'>
print(add(5, 3)) # Output: 8
```

```python
even = lambda num: num%2==0
print(even(4)) # Output: True
```
